# infixToPostfix() 

아래 코드는 우리가 Infix, Prefix, Postfix를 배우면서 구현한 infixToPostfix() 함수입니다. 

````python
#%%writefile infixToPostfix.py
from dspy import Stack

def infixToPostfix(infixexpr):
    pred = {}
    pred["*"] = 3
    pred["/"] = 3
    pred["+"] = 2
    pred["-"] = 2
    pred["("] = 1
    opStack = Stack()
    postfixList = []
    tokenList = infixexpr.split()

    for token in tokenList:
        if token in 'ABCDEFGHIJKLMNOPQRSTUVWXYZ' or token in "0123456789":
            postfixList.append(token)
        elif token == '(':
            opStack.push(token)
        elif token == ')':
            topToken = opStack.pop()
            while topToken != '(':
                postfixList.append(topToken)
                topToken = opStack.pop()
        else:
            while (not opStack.is_empty()) and \
               (pred[opStack.peek()] >= pred[token]):
                  postfixList.append(opStack.pop())
            opStack.push(token)

    while not opStack.is_empty():
        postfixList.append(opStack.pop())
    return " ".join(postfixList)

if __name__ == "__main__":
    print(infixToPostfix("( A + B ) * C"))
    print(infixToPostfix("A + B * C"))
    print(infixToPostfix("A * B + C * D"))
    print(infixToPostfix("( A + B ) * C - ( D - E ) * ( F + G )"))
    print(infixToPostfix("1 + 2 * ( 3 + 4 )"))
````

위의 기술된 `infixToPostfix()` 함수는 사용면에서 여러가지 제약점이 있습니다. 

1. 수식(`infixexpr`)이 한 자(one uppercase letter or one digit)로만 구성된 연산자와 피연산자로 표현되어야 합니다.  
    예를 들어, `a + bc, ab * c, 1 + 23` 같은 수식이 가능하지 않습니다. 
1. 연산자와 피연산자 사이에 빈칸이 있어야 합니다.   
    예를 들어, `1-2+3` 같은 수식이 가능하지 않습니다. 
1. 정수만 사용해야 하고, 소숫점을 허용하지 않습니다.  
    예를 들어, `1.23 + 44` 같은 수식이 가능하지 않습니다. 
1. 제곱(^) 연산자를 허용하지 않습니다.   
    예를 들어, `2^10 + 5` 같은 수식이 가능하지 않습니다.  

위와 같은 제약없이 표현된 infix 수식을 postfix로 전환하는 함수를 구현하고자 합니다. 단, 위의 `infixToPostfix()` 함수를 기본으로 수정해나가도록 합니다. 

또한 `postEval()` 함수도 이러한 제약들 없이 연산이 가능하도록 수정합니다.

````python
from dspy import Stack

def postfixEval(postfixExpr):
    operandStack = Stack()
    tokenList = postfixExpr.split()            # you may keep this line or replace it if necessary

    for token in tokenList:
        if token in "0123456789":
            operandStack.push(int(token))
        else:
            operand2 = operandStack.pop()
            operand1 = operandStack.pop()
            result = doMath(token,operand1,operand2)
            operandStack.push(result)
    return operandStack.pop()

def doMath(op, op1, op2):
    if op == "*":
        return op1 * op2
    elif op == "/":
        return op1 / op2
    elif op == "+":
        return op1 + op2
    elif op == "-":
        return op1 - op2
    else: 
        print("cannot process the operator: " + str(op))

if __name__ == "__main__":
    print(postfixEval('7 8 + 3 2 + /'))
    print(postfixEval('1 2 3 4 + * +'))
````
